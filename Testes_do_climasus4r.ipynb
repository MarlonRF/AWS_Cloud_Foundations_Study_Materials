# Código das funções

````r
# Instale se necessário
install.packages(c(
  "shiny", "plotly","IRdisplay", "ggplot2", "dplyr", "tidyr", "lubridate"
))

library(shiny)
library(plotly)
library(ggplot2)
library(dplyr)
library(tidyr)
library(lubridate)
library(IRdisplay)

app_graficos_interativos <- function(df) {

  d <- as.data.frame(df)
  names(d) <- make.names(names(d), unique = TRUE)

  num_cols <- names(d)[vapply(d, is.numeric, logical(1))]

  cat_cols <- names(d)[vapply(
    d,
    function(x) is.character(x) || is.factor(x) || is.logical(x),
    logical(1)
  )]

  all_cols <- names(d)

  date_cols <- names(d)[vapply(
    d,
    function(x) inherits(x, c("Date", "POSIXct", "POSIXlt")),
    logical(1)
  )]

  # tenta detectar colunas de data em texto
  text_date_cols <- names(d)[vapply(
    d,
    function(x) {
      if (!(is.character(x) || is.factor(x))) return(FALSE)
      x2 <- suppressWarnings(as.POSIXct(as.character(x), tz = "UTC"))
      mean(!is.na(x2)) > 0.5
    },
    logical(1)
  )]

  date_cols <- unique(c(date_cols, text_date_cols))

  if (length(num_cols) == 0) {
    stop("O data.frame precisa ter pelo menos uma coluna numérica.")
  }

  if (length(num_cols) == 1) {
    num_cols_y <- num_cols
  } else {
    num_cols_y <- num_cols[-1]
  }

  cat_options <- c("None", cat_cols)
  date_options <- if (length(date_cols) > 0) date_cols else all_cols

  log1p_seguro <- function(x) {
    ifelse(is.na(x) | x <= -1, NA_real_, log1p(x))
  }

  agg_fun <- function(x, metodo) {
    switch(
      metodo,
      mean = mean(x, na.rm = TRUE),
      sum = sum(x, na.rm = TRUE),
      median = median(x, na.rm = TRUE),
      min = min(x, na.rm = TRUE),
      max = max(x, na.rm = TRUE)
    )
  }

  ui <- fluidPage(
    titlePanel("Gráficos interativos em R"),

    sidebarLayout(
      sidebarPanel(
        selectInput(
          "tipo",
          "Tipo de gráfico",
          choices = c(
            "Dispersão",
            "Histograma",
            "Linha",
            "Série temporal",
            "Boxplot",
            "Mapa de calor - correlação",
            "Mapa de calor - agregado"
          )
        ),

        uiOutput("x_ui"),
        uiOutput("y_ui"),
        uiOutput("z_ui"),
        uiOutput("cor_ui"),
        uiOutput("tamanho_ui"),

        conditionalPanel(
          condition = "input.tipo == 'Histograma'",
          sliderInput("bins", "Bins", min = 5, max = 100, value = 30, step = 1)
        ),

        conditionalPanel(
          condition = "input.tipo == 'Série temporal'",
          selectInput(
            "freq",
            "Frequência",
            choices = c(
              "nenhuma",
              "dia",
              "semana",
              "mes",
              "trimestre",
              "ano"
            ),
            selected = "nenhuma"
          )
        ),

        conditionalPanel(
          condition = "
            input.tipo == 'Série temporal' ||
            input.tipo == 'Mapa de calor - agregado'
          ",
          selectInput(
            "agreg",
            "Agregação",
            choices = c("mean", "sum", "median", "min", "max", "count"),
            selected = "mean"
          )
        ),

        conditionalPanel(
          condition = "
            input.tipo == 'Dispersão' ||
            input.tipo == 'Linha' ||
            input.tipo == 'Histograma'
          ",
          checkboxInput("log_x", "log(X + 1)", value = FALSE)
        ),

        conditionalPanel(
          condition = "
            input.tipo == 'Dispersão' ||
            input.tipo == 'Linha' ||
            input.tipo == 'Série temporal' ||
            input.tipo == 'Boxplot'
          ",
          checkboxInput("log_y", "log(Y + 1)", value = FALSE)
        ),

        sliderInput(
          "opacidade",
          "Opacidade",
          min = 0.1,
          max = 1,
          value = 0.7,
          step = 0.05
        ),

        checkboxInput("excluir_na", "Excluir NA", value = TRUE)
      ),

      mainPanel(
        plotlyOutput("grafico", height = "650px")
      )
    )
  )

  server <- function(input, output, session) {

    output$x_ui <- renderUI({
      req(input$tipo)

      if (input$tipo == "Histograma") {
        selectInput("x", "Variável", choices = num_cols, selected = num_cols[1])
      } else if (input$tipo == "Série temporal") {
        selectInput("x", "Data", choices = date_options, selected = date_options[1])
      } else if (input$tipo == "Boxplot") {
        selectInput("x", "Grupo X", choices = cat_options, selected = "None")
      } else if (input$tipo == "Mapa de calor - agregado") {
        selectInput("x", "X", choices = all_cols, selected = all_cols[1])
      } else if (input$tipo %in% c("Dispersão", "Linha")) {
        selectInput("x", "X", choices = num_cols, selected = num_cols[1])
      }
    })

    output$y_ui <- renderUI({
      req(input$tipo)

      if (input$tipo %in% c("Dispersão", "Linha", "Série temporal", "Boxplot")) {
        selectInput("y", "Y", choices = num_cols, selected = num_cols[length(num_cols)])
      } else if (input$tipo == "Mapa de calor - agregado") {
        selectInput("y", "Y", choices = all_cols, selected = all_cols[min(2, length(all_cols))])
      }
    })

    output$z_ui <- renderUI({
      req(input$tipo)

      if (input$tipo == "Mapa de calor - agregado") {
        selectInput("z", "Valor", choices = num_cols, selected = num_cols[1])
      }
    })

    output$cor_ui <- renderUI({
      req(input$tipo)

      if (input$tipo %in% c(
        "Dispersão", "Histograma", "Linha", "Série temporal", "Boxplot"
      )) {
        selectInput("cor", "Cor", choices = cat_options, selected = "None")
      }
    })

    output$tamanho_ui <- renderUI({
      req(input$tipo)

      if (input$tipo == "Dispersão") {
        selectInput("tamanho", "Tamanho", choices = c("None", num_cols), selected = "None")
      }
    })

    limpar_na <- function(dd, cols) {
      cols <- cols[cols != "None"]
      cols <- cols[!is.na(cols)]

      if (isTRUE(input$excluir_na) && length(cols) > 0) {
        dd <- tidyr::drop_na(dd, dplyr::all_of(cols))
      }

      dd
    }

    output$grafico <- renderPlotly({
      req(input$tipo)

      dd <- d

      if (input$tipo == "Dispersão") {
        req(input$x, input$y)

        cols <- c(input$x, input$y, input$cor, input$tamanho)
        dd <- limpar_na(dd, cols)

        x_plot <- input$x
        y_plot <- input$y
        x_lab <- input$x
        y_lab <- input$y

        if (isTRUE(input$log_x)) {
          x_plot <- paste0(input$x, "__log1p")
          dd[[x_plot]] <- log1p_seguro(dd[[input$x]])
          x_lab <- paste0("log(", input$x, " + 1)")
        }

        if (isTRUE(input$log_y)) {
          y_plot <- paste0(input$y, "__log1p")
          dd[[y_plot]] <- log1p_seguro(dd[[input$y]])
          y_lab <- paste0("log(", input$y, " + 1)")
        }

        g <- ggplot(
          dd,
          aes(
            x = .data[[x_plot]],
            y = .data[[y_plot]],
            text = paste0(
              x_lab, ": ", .data[[x_plot]],
              "<br>", y_lab, ": ", .data[[y_plot]]
            )
          )
        )

        if (input$cor != "None" && input$tamanho != "None") {
          g <- g + geom_point(
            aes(color = .data[[input$cor]], size = .data[[input$tamanho]]),
            alpha = input$opacidade
          )
        } else if (input$cor != "None") {
          g <- g + geom_point(
            aes(color = .data[[input$cor]]),
            alpha = input$opacidade
          )
        } else if (input$tamanho != "None") {
          g <- g + geom_point(
            aes(size = .data[[input$tamanho]]),
            alpha = input$opacidade
          )
        } else {
          g <- g + geom_point(alpha = input$opacidade)
        }

        g <- g +
          labs(
            title = paste("Dispersão:", y_lab, "vs", x_lab),
            x = x_lab,
            y = y_lab
          ) +
          theme_minimal()

        return(ggplotly(g, tooltip = "text"))
      }

      if (input$tipo == "Histograma") {
        req(input$x)

        cols <- c(input$x, input$cor)
        dd <- limpar_na(dd, cols)

        x_plot <- input$x
        x_lab <- input$x

        if (isTRUE(input$log_x)) {
          x_plot <- paste0(input$x, "__log1p")
          dd[[x_plot]] <- log1p_seguro(dd[[input$x]])
          x_lab <- paste0("log(", input$x, " + 1)")
        }

        g <- ggplot(dd, aes(x = .data[[x_plot]]))

        if (input$cor != "None") {
          g <- g + geom_histogram(
            aes(fill = .data[[input$cor]]),
            bins = input$bins,
            alpha = input$opacidade,
            position = "identity"
          )
        } else {
          g <- g + geom_histogram(
            bins = input$bins,
            alpha = input$opacidade
          )
        }

        g <- g +
          labs(
            title = paste("Histograma de", x_lab),
            x = x_lab,
            y = "Contagem"
          ) +
          theme_minimal()

        return(ggplotly(g))
      }

      if (input$tipo == "Linha") {
        req(input$x, input$y)

        cols <- c(input$x, input$y, input$cor)
        dd <- limpar_na(dd, cols)

        x_plot <- input$x
        y_plot <- input$y
        x_lab <- input$x
        y_lab <- input$y

        if (isTRUE(input$log_x)) {
          x_plot <- paste0(input$x, "__log1p")
          dd[[x_plot]] <- log1p_seguro(dd[[input$x]])
          x_lab <- paste0("log(", input$x, " + 1)")
        }

        if (isTRUE(input$log_y)) {
          y_plot <- paste0(input$y, "__log1p")
          dd[[y_plot]] <- log1p_seguro(dd[[input$y]])
          y_lab <- paste0("log(", input$y, " + 1)")
        }

        dd <- dd |> arrange(.data[[x_plot]])

        if (input$cor != "None") {
          g <- ggplot(
            dd,
            aes(
              x = .data[[x_plot]],
              y = .data[[y_plot]],
              color = .data[[input$cor]],
              text = paste0(
                x_lab, ": ", .data[[x_plot]],
                "<br>", y_lab, ": ", .data[[y_plot]]
              )
            )
          ) +
            geom_line(alpha = input$opacidade) +
            geom_point(alpha = input$opacidade)
        } else {
          g <- ggplot(
            dd,
            aes(
              x = .data[[x_plot]],
              y = .data[[y_plot]],
              group = 1,
              text = paste0(
                x_lab, ": ", .data[[x_plot]],
                "<br>", y_lab, ": ", .data[[y_plot]]
              )
            )
          ) +
            geom_line(alpha = input$opacidade) +
            geom_point(alpha = input$opacidade)
        }

        g <- g +
          labs(
            title = paste("Linha:", y_lab, "vs", x_lab),
            x = x_lab,
            y = y_lab
          ) +
          theme_minimal()

        return(ggplotly(g, tooltip = "text"))
      }

      if (input$tipo == "Série temporal") {
        req(input$x, input$y)

        dd[[input$x]] <- suppressWarnings(as.POSIXct(as.character(dd[[input$x]]), tz = "UTC"))

        cols <- c(input$x, input$y, input$cor)
        dd <- limpar_na(dd, cols)

        y_plot <- input$y
        y_lab <- input$y

        if (isTRUE(input$log_y)) {
          y_plot <- paste0(input$y, "__log1p")
          dd[[y_plot]] <- log1p_seguro(dd[[input$y]])
          y_lab <- paste0("log(", input$y, " + 1)")
        }

        x_final <- input$x
        y_final <- y_plot

        if (input$freq != "nenhuma") {
          unidade <- switch(
            input$freq,
            dia = "day",
            semana = "week",
            mes = "month",
            trimestre = "quarter",
            ano = "year"
          )

          dd$periodo <- lubridate::floor_date(dd[[input$x]], unit = unidade)

          keys <- c("periodo")
          if (input$cor != "None") keys <- c(keys, input$cor)

          if (input$agreg == "count") {
            dd <- dd |>
              group_by(across(all_of(keys))) |>
              summarise(valor = n(), .groups = "drop")
          } else {
            dd <- dd |>
              group_by(across(all_of(keys))) |>
              summarise(valor = agg_fun(.data[[y_plot]], input$agreg), .groups = "drop")
          }

          x_final <- "periodo"
          y_final <- "valor"
        } else if (input$agreg == "count") {
          keys <- c(input$x)
          if (input$cor != "None") keys <- c(keys, input$cor)

          dd <- dd |>
            group_by(across(all_of(keys))) |>
            summarise(valor = n(), .groups = "drop")

          y_final <- "valor"
        }

        dd <- dd |> arrange(.data[[x_final]])

        if (input$cor != "None") {
          g <- ggplot(
            dd,
            aes(
              x = .data[[x_final]],
              y = .data[[y_final]],
              color = .data[[input$cor]],
              text = paste0(
                "Data: ", .data[[x_final]],
                "<br>Valor: ", .data[[y_final]]
              )
            )
          ) +
            geom_line(alpha = input$opacidade) +
            geom_point(alpha = input$opacidade)
        } else {
          g <- ggplot(
            dd,
            aes(
              x = .data[[x_final]],
              y = .data[[y_final]],
              group = 1,
              text = paste0(
                "Data: ", .data[[x_final]],
                "<br>Valor: ", .data[[y_final]]
              )
            )
          ) +
            geom_line(alpha = input$opacidade) +
            geom_point(alpha = input$opacidade)
        }

        g <- g +
          labs(
            title = paste("Série temporal:", y_lab),
            x = "Data",
            y = ifelse(input$agreg == "count", "Contagem", y_lab)
          ) +
          theme_minimal()

        return(ggplotly(g, tooltip = "text"))
      }

      if (input$tipo == "Boxplot") {
        req(input$y)

        cols <- c(input$y, input$x, input$cor)
        dd <- limpar_na(dd, cols)

        y_plot <- input$y
        y_lab <- input$y

        if (isTRUE(input$log_y)) {
          y_plot <- paste0(input$y, "__log1p")
          dd[[y_plot]] <- log1p_seguro(dd[[input$y]])
          y_lab <- paste0("log(", input$y, " + 1)")
        }

        if (input$x != "None" && input$cor != "None") {
          g <- ggplot(
            dd,
            aes(
              x = .data[[input$x]],
              y = .data[[y_plot]],
              fill = .data[[input$cor]]
            )
          ) +
            geom_boxplot(alpha = input$opacidade)
        } else if (input$x != "None") {
          g <- ggplot(
            dd,
            aes(
              x = .data[[input$x]],
              y = .data[[y_plot]]
            )
          ) +
            geom_boxplot(alpha = input$opacidade)
        } else if (input$cor != "None") {
          g <- ggplot(
            dd,
            aes(
              x = .data[[input$cor]],
              y = .data[[y_plot]],
              fill = .data[[input$cor]]
            )
          ) +
            geom_boxplot(alpha = input$opacidade)
        } else {
          g <- ggplot(
            dd,
            aes(
              x = "",
              y = .data[[y_plot]]
            )
          ) +
            geom_boxplot(alpha = input$opacidade)
        }

        g <- g +
          labs(
            title = paste("Boxplot de", y_lab),
            x = "",
            y = y_lab
          ) +
          theme_minimal()

        return(ggplotly(g))
      }

      if (input$tipo == "Mapa de calor - correlação") {
        req(length(num_cols) >= 2)

        base <- dd[, num_cols, drop = FALSE]

        if (isTRUE(input$excluir_na)) {
          base <- tidyr::drop_na(base)
        }

        corr <- cor(base, use = "pairwise.complete.obs")

        corr_long <- as.data.frame(as.table(corr))
        names(corr_long) <- c("linha", "coluna", "correlacao")

        g <- ggplot(
          corr_long,
          aes(
            x = coluna,
            y = linha,
            fill = correlacao,
            text = paste0(
              "Linha: ", linha,
              "<br>Coluna: ", coluna,
              "<br>Correlação: ", round(correlacao, 3)
            )
          )
        ) +
          geom_tile() +
          coord_equal() +
          labs(
            title = "Mapa de calor da correlação",
            x = "",
            y = "",
            fill = "Correlação"
          ) +
          theme_minimal()

        return(ggplotly(g, tooltip = "text"))
      }

      if (input$tipo == "Mapa de calor - agregado") {
        req(input$x, input$y, input$z)

        cols <- c(input$x, input$y)
        if (input$agreg != "count") cols <- c(cols, input$z)

        dd <- limpar_na(dd, cols)

        if (input$agreg == "count") {
          tab <- dd |>
            group_by(.data[[input$x]], .data[[input$y]]) |>
            summarise(valor = n(), .groups = "drop")
        } else {
          tab <- dd |>
            group_by(.data[[input$x]], .data[[input$y]]) |>
            summarise(valor = agg_fun(.data[[input$z]], input$agreg), .groups = "drop")
        }

        g <- ggplot(
          tab,
          aes(
            x = .data[[input$x]],
            y = .data[[input$y]],
            fill = valor,
            text = paste0(
              input$x, ": ", .data[[input$x]],
              "<br>", input$y, ": ", .data[[input$y]],
              "<br>Valor: ", round(valor, 3)
            )
          )
        ) +
          geom_tile() +
          labs(
            title = paste("Mapa de calor:", input$agreg),
            x = input$x,
            y = input$y,
            fill = "Valor"
          ) +
          theme_minimal()

        return(ggplotly(g, tooltip = "text"))
      }
    })
  }

  shinyApp(ui, server)
}
````
